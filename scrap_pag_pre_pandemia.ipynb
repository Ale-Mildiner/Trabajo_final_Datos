#Departamentos

```python
#agarramos los nombres de deptos y sus links

#Aca hay que poner el link del cuatrimestre

response = requests.get('https://encuestas-finales.exactas.uba.ar/cuat/u22.html')
soup = BS(response.content)
soup.prettify

elements = soup.find_all(attrs = {"class": "list"})
links = elements[2].find_all('a')
deptos = []
enlaces = []
for i in range(len(links)):
  enlaces.append('https://encuestas-finales.exactas.uba.ar'+links[i]['href'][2:])
  deptos.append(str(links[i]).replace((links[i]['href']),'')[11:-4])
```

```python
deptos
```

#Materias

```python
departamentos = []
materias = []

for i in range(len(enlaces[:14])):
  response = requests.get(enlaces[i])
  soup = BS(response.content)
  browser.get(enlaces[i])
  listas = soup.find_all(attrs = {"class": "list"})
  if len(listas) !=0:
    listapags = listas[0].find_all(attrs = {"class": "head"})#veo cuantas pags hay
    if len(listapags) !=0:                                #si la lista tiene 0 elementos hay 1 sola
      pags = int(str(listapags[0].find_all('a')[-2])[-5])
      listamaterias = soup.find_all(attrs = {"class": "noprint"}) #scrapeo la 1er pag
      for j in range(len(listamaterias)):  
        materias.append('https://encuestas-finales.exactas.uba.ar/mat'+listamaterias[j]['href'][6:-8]+'.html')
        departamentos.append(deptos[i])
      for k in range(pags-1):          #paso de pag
        sig = browser.find_element_by_partial_link_text('Siguiente')
        sig.click()
        body = browser.execute_script("return document.body")
        source = body.get_attribute('innerHTML') 
        soup = BS(source, "html.parser")
        listamaterias = soup.find_all(attrs = {"class": "noprint"})  #scrapeo las proxs pags
        for m in range(len(listamaterias)):
          materias.append('https://encuestas-finales.exactas.uba.ar/mat'+listamaterias[m]['href'][6:-8]+'.html')
          departamentos.append(deptos[i])
    else:
      listamaterias = soup.find_all(attrs = {"class": "noprint"})  #scrapeo las proxs pags  
      for j in range(len(listamaterias)):
        materias.append('https://encuestas-finales.exactas.uba.ar/mat'+listamaterias[j]['href'][6:-8]+'.html')
        departamentos.append(deptos[i])
```
